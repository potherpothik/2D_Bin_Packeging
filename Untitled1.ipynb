```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


class RectanglePacker:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.spaces = [(0, 0, width, height)]  # List of available spaces
    
    def pack(self, rect_width, rect_height):
        for i, (x, y, w, h) in enumerate(self.spaces):
            if w >= rect_width and h >= rect_height:  # Check normal orientation
                return self._place_rectangle(i, x, y, rect_width, rect_height, w, h)
            elif w >= rect_height and h >= rect_width:  # Check rotated orientation
                return self._place_rectangle(i, x, y, rect_height, rect_width, w, h)
        raise ValueError("Cannot pack rectangle")  # No space found
    
    def _place_rectangle(self, index, x, y, rect_width, rect_height, space_width, space_height):
        # Remove the used space
        self.spaces.pop(index)
        
        # Add new spaces based on remaining areas
        if space_width > rect_width:
            self.spaces.append((x + rect_width, y, space_width - rect_width, rect_height))
        if space_height > rect_height:
            self.spaces.append((x, y + rect_height, rect_width, space_height - rect_height))
        
        # Return the coordinates where the rectangle was placed
        return x, y

class GlassCuttingOptimizer:
    def __init__(self, glass_data_file, stock_sizes_file):
        self.glass_parts = pd.read_csv(glass_data_file)
        self.stock_sizes = pd.read_csv(stock_sizes_file).drop_duplicates(subset=['length', 'width'])
        self.expanded_parts = self.prepare_parts()
    
    def prepare_parts(self):
        expanded_parts = []
        for _, part in self.glass_parts.iterrows():
            expanded_parts.extend([{
                'location': part['location'],
                'length': part['glass_length'],
                'height': part['glass_height']
            } for _ in range(part['glass_qty'])])
        # Sort by largest dimension (longer side first)
        return sorted(expanded_parts, key=lambda x: max(x['length'], x['height']), reverse=True)
    
    def optimize_cutting(self, buffer=0):
        total_parts_area = sum((part['length'] + buffer) * (part['height'] + buffer) for part in self.expanded_parts)
    
        stock_sizes = self.stock_sizes.sort_values(by=['length', 'width'])
        best_stock_size = self.select_optimal_stock_size(total_parts_area, buffer)
        sheet_length = best_stock_size['length']
        sheet_width = best_stock_size['width']
    
        packed_parts = []
        remaining_parts = self.expanded_parts.copy()
        total_sheets_used = 0
    
        while remaining_parts:
            packer = RectanglePacker(sheet_length, sheet_width)
            current_sheet_parts = []
    
            for part in remaining_parts[:]:
                try:
                    x, y = packer.pack(part['length'] + buffer, part['height'] + buffer)
                    current_sheet_parts.append({
                        **part,
                        'x': x,
                        'y': y,
                        'packed_length': part['length'],
                        'packed_height': part['height']
                    })
                    remaining_parts.remove(part)
                except ValueError:
                    continue
    
            if not current_sheet_parts:
                raise ValueError("Not enough stock sheets to fit all parts. Check your stock sizes.")
    
            packed_parts.extend(current_sheet_parts)
            total_sheets_used += 1
    
        total_stock_area = total_sheets_used * (sheet_length * sheet_width)
        used_area_percentage = (total_parts_area / total_stock_area) * 100
        wastage_percentage = 100 - used_area_percentage
    
        return {
            'total_glass_area': total_parts_area / 1_000_000,
            'total_stock_area': total_stock_area / 1_000_000,
            'used_area_percentage': used_area_percentage,
            'wastage_percentage': wastage_percentage,
            'total_sheets_used': total_sheets_used,
            'stock_size': best_stock_size,
            'packed_parts': packed_parts
        }

 
    def select_optimal_stock_size(self, total_parts_area, buffer):
        stock_sizes = self.stock_sizes.sort_values(by=['length', 'width'])
        for _, size in stock_sizes.iterrows():
            if size['length'] * size['width'] >= total_parts_area * buffer:
                return size
        return stock_sizes.iloc[-1]  # Return the largest stock if no fit found
    
    def pack_rectangles(self, stock_size):
        packer = RectanglePacker(stock_size['length'], stock_size['width'])
        packed_parts = []
        failed_parts = []
        for part in self.expanded_parts:
            try:
                x, y = packer.pack(part['length'], part['height'])
                packed_parts.append({
                    **part,
                    'x': x,
                    'y': y
                })
            except ValueError:
                failed_parts.append(part)
        return packed_parts, failed_parts

# Main execution
optimizer = GlassCuttingOptimizer(
    glass_data_file='data/glass_data.csv', 
    stock_sizes_file='data/glass_sheet_size.csv'
)

result = optimizer.optimize_cutting()


# Print key metrics
print(f"Total glass area: {result['total_glass_area']:.3f} sq m")
print(f"Total stock area used: {result['total_stock_area']:.3f} sq m")
print(f"Used area percentage: {result['used_area_percentage']:.2f}%")
print(f"Wastage percentage: {result['wastage_percentage']:.2f}%")
print(f"Failed parts: {len(result['failed_parts'])}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class GlassCuttingOptimizer:
    def __init__(self, glass_data_file, stock_sizes_file):
        # Load data
        self.glass_parts = pd.read_csv(glass_data_file)
        self.stock_sizes = pd.read_csv(stock_sizes_file)
        
        # Preprocess and expand parts
        self.expanded_parts = self.prepare_parts()
    
    def prepare_parts(self):
        # Expand parts based on quantity
        expanded_parts = []
        for _, part in self.glass_parts.iterrows():
            expanded_parts.extend([{
                'location': part['location'],
                'length': part['glass_length'],
                'height': part['glass_height']
            } for _ in range(part['glass_qty'])])
        
        # Sort parts by area in descending order
        return sorted(expanded_parts, key=lambda x: x['length'] * x['height'], reverse=True)
    
    def optimize_cutting(self):
        # Calculate total parts area
        total_parts_area = sum(part['length'] * part['height'] for part in self.expanded_parts)
        
        # Determine optimal stock sizes
        total_stock_area = total_parts_area * 1.2  # 20% overhead
        
        # Select appropriate stock sizes
        sheets_needed = self.calculate_sheets_needed(total_stock_area)
        
        # Detailed calculations
        actual_stock_area = sum(sheet['length'] * sheet['width'] for sheet in sheets_needed)
        used_area_percentage = (total_parts_area / actual_stock_area) * 100
        wastage_percentage = 100 - used_area_percentage
        
        return {
            'total_glass_area': total_parts_area / 1_000_000,  # sq m
            'total_stock_area': actual_stock_area / 1_000_000,  # sq m
            'used_area_percentage': used_area_percentage,
            'wastage_percentage': wastage_percentage,
            'sheets_used': sheets_needed
        }
    
    def calculate_sheets_needed(self, total_required_area):
        # Sort stock sizes by area
        sorted_sizes = self.stock_sizes.sort_values(by=['length', 'width'])
        
        # Select appropriate sheets
        sheets_needed = []
        remaining_area = total_required_area
        
        for _, sheet in sorted_sizes.iterrows():
            sheet_area = sheet['length'] * sheet['width']
            
            while remaining_area > 0 and sheet_area <= remaining_area:
                sheets_needed.append(sheet.to_dict())
                remaining_area -= sheet_area
        
        return sheets_needed
    
    def visualize_results(self, optimization_result):
        plt.figure(figsize=(16, 10))
        
        # Sheet Size Distribution
        plt.subplot(2, 2, 1)
        self.plot_sheet_size_distribution(optimization_result)
        
        # Performance Metrics
        plt.subplot(2, 2, 2)
        self.plot_performance_metrics(optimization_result)
        
        # Part Size Distribution
        plt.subplot(2, 2, 3)
        self.plot_part_distribution()
        
        # Stock Size Utilization
        plt.subplot(2, 2, 4)
        self.plot_stock_size_usage(optimization_result)
        
        plt.tight_layout()
        plt.show()
    
    def plot_sheet_size_distribution(self, optimization_result):
        sheets_used = optimization_result['sheets_used']
        sizes_df = pd.DataFrame(sheets_used)
        
        plt.scatter(sizes_df['length'], sizes_df['width'])
        plt.title("Used Stock Sizes")
        plt.xlabel("Length (mm)")
        plt.ylabel("Width (mm)")
    
    def plot_performance_metrics(self, optimization_result):
        metrics = {
            'Total Glass Area (sq m)': optimization_result['total_glass_area'],
            'Total Stock Area (sq m)': optimization_result['total_stock_area'],
            'Used Area (%)': optimization_result['used_area_percentage'],
            'Wastage (%)': optimization_result['wastage_percentage']
        }
        
        plt.bar(metrics.keys(), metrics.values())
        plt.title("Cutting Performance Metrics")
        plt.xticks(rotation=45, ha='right')
    
    def plot_part_distribution(self):
        lengths = [part['length'] for part in self.expanded_parts]
        heights = [part['height'] for part in self.expanded_parts]
        
        plt.scatter(lengths, heights)
        plt.title("Part Size Distribution")
        plt.xlabel("Length (mm)")
        plt.ylabel("Height (mm)")
    
    def plot_stock_size_usage(self, optimization_result):
        sheets_used = optimization_result['sheets_used']
        
        sheet_counts = {}
        for sheet in sheets_used:
            key = (sheet['length'], sheet['width'])
            sheet_counts[key] = sheet_counts.get(key, 0) + 1
        
        plt.bar([f"{l}x{w}" for (l,w) in sheet_counts.keys()], 
                list(sheet_counts.values()))
        plt.title("Stock Size Usage")
        plt.xlabel("Sheet Size (mm)")
        plt.ylabel("Number of Sheets")
        plt.xticks(rotation=45, ha='right')

# Main execution
optimizer = GlassCuttingOptimizer(
    glass_data_file='data/glass_data.csv', 
    stock_sizes_file='data/glass_sheet_size.csv'
)

# Perform optimization
result = optimizer.optimize_cutting()

# Print key metrics
print(f"Total glass area: {result['total_glass_area']:.3f} sq m")
print(f"Total stock area used: {result['total_stock_area']:.3f} sq m")
print(f"Used area percentage: {result['used_area_percentage']:.2f}%")
print(f"Wastage percentage: {result['wastage_percentage']:.2f}%")

# Visualize results
optimizer.visualize_results(result)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dataclasses import dataclass, field
from typing import List, Tuple

@dataclass
class Part:
    location: str
    length: int
    height: int

@dataclass
class Placement:
    part: Part
    x: int
    y: int
    rotated: bool = False

@dataclass
class Sheet:
    length: int
    width: int
    placements: List[Placement] = field(default_factory=list)

class GlassCuttingOptimizer:
    def __init__(self, glass_data_file, stock_sizes_file):
        # Load data
        self.glass_parts = pd.read_csv(glass_data_file)
        self.stock_sizes = pd.read_csv(stock_sizes_file)
        
        # Preprocess and expand parts
        self.expanded_parts = self.prepare_parts()
    
    def prepare_parts(self):
        # Expand parts based on quantity
        expanded_parts = []
        for _, part in self.glass_parts.iterrows():
            expanded_parts.extend([Part(
                location=part['location'],
                length=part['glass_length'],
                height=part['glass_height']
            ) for _ in range(part['glass_qty'])])
        
        # Sort parts by area in descending order
        return sorted(expanded_parts, key=lambda x: x.length * x.height, reverse=True)
    
    def optimize_cutting(self, gap=5):
        total_parts_area = sum(part.length * part.height for part in self.expanded_parts)
        
        # Calculate required area with 20% overhead
        total_stock_area = total_parts_area * 1.2
        
        # Prepare sheets and layout
        sheets = self.calculate_layout(self.expanded_parts, total_stock_area, gap)
        
        # Calculate metrics
        actual_stock_area = sum(sheet.length * sheet.width for sheet in sheets)
        used_area_percentage = (total_parts_area / actual_stock_area) * 100
        wastage_percentage = 100 - used_area_percentage
        
        return {
            'total_glass_area': total_parts_area / 1_000_000,  # sq m
            'total_stock_area': actual_stock_area / 1_000_000,  # sq m
            'used_area_percentage': used_area_percentage,
            'wastage_percentage': wastage_percentage,
            'sheets': sheets
        }
    
    def calculate_layout(self, parts, total_stock_area, gap):
        # Simplified layout calculation
        sheets = []
        remaining_parts = parts.copy()
        
        while remaining_parts:
            # Select appropriate sheet size
            current_sheet = self.select_sheet_size(total_stock_area)
            
            # Place parts on the sheet
            sheet = Sheet(current_sheet['length'], current_sheet['width'])
            
            for part in remaining_parts[:]:
                placement = self.place_part(sheet, part, gap)
                if placement:
                    sheet.placements.append(placement)
                    remaining_parts.remove(part)
            
            sheets.append(sheet)
        
        return sheets
    
    def select_sheet_size(self, total_stock_area):
        # Select sheet size based on required area
        sorted_sizes = self.stock_sizes.sort_values(by=['length', 'width'])
        for _, sheet in sorted_sizes.iterrows():
            if sheet['length'] * sheet['width'] >= total_stock_area:
                return sheet
        
        # Fallback to largest sheet
        return sorted_sizes.iloc[-1]
    
    def place_part(self, sheet, part, gap):
        # Basic placement strategy
        for x in range(0, sheet.length - part.length, gap):
            for y in range(0, sheet.width - part.height, gap):
                # Check if placement is possible
                if self.is_valid_placement(sheet, part, x, y):
                    return Placement(part, x, y)
                
                # Try rotated placement
                if self.is_valid_placement(sheet, part, x, y, rotated=True):
                    return Placement(part, x, y, rotated=True)
        
        return None
    
    def is_valid_placement(self, sheet, part, x, y, rotated=False):
        # Check placement validity
        if rotated:
            width, height = part.height, part.length
        else:
            width, height = part.length, part.height
        
        return (x + width <= sheet.length and 
                y + height <= sheet.width)
    
    def print_layout_report(self, optimization_result):
        print("\nOptimized Layout:")
        for i, sheet in enumerate(optimization_result['sheets'], 1):
            print(f"Sheet {i}: {sheet.length}x{sheet.width}")
            for placement in sheet.placements:
                orientation = "rotated" if placement.rotated else "normal"
                print(f"  {placement.part.location} ({placement.part.length}x{placement.part.height}) at position ({placement.x},{placement.y}) ({orientation})")
    
    def visualize_layout(self, optimization_result):
        plt.figure(figsize=(20, 10))
        
        for i, sheet in enumerate(optimization_result['sheets'], 1):
            plt.subplot(1, len(optimization_result['sheets']), i)
            plt.title(f"Sheet {i}: {sheet.length}x{sheet.width}")
            plt.gca().add_patch(plt.Rectangle((0, 0), sheet.length, sheet.width, fill=False, edgecolor='black'))
            
            for placement in sheet.placements:
                x, y = placement.x, placement.y
                if placement.rotated:
                    width, height = placement.part.height, placement.part.length
                else:
                    width, height = placement.part.length, placement.part.height
                
                plt.gca().add_patch(plt.Rectangle(
                    (x, y), width, height, 
                    fill=True, alpha=0.3, 
                    label=f"{placement.part.location}"
                ))
                plt.text(x + width/2, y + height/2, placement.part.location, 
                         horizontalalignment='center', verticalalignment='center')
            
            plt.xlabel("Length (mm)")
            plt.ylabel("Width (mm)")
            plt.xlim(0, sheet.length)
            plt.ylim(0, sheet.width)
        
        plt.tight_layout()
        plt.show()

# Main execution
optimizer = GlassCuttingOptimizer(
    glass_data_file='data/glass_data.csv', 
    stock_sizes_file='data/glass_sheet_size.csv'
)

# Perform optimization
result = optimizer.optimize_cutting()

# Print key metrics
print(f"Total glass area: {result['total_glass_area']:.3f} sq m")
print(f"Total stock area used: {result['total_stock_area']:.3f} sq m")
print(f"Used area percentage: {result['used_area_percentage']:.2f}%")
print(f"Wastage percentage: {result['wastage_percentage']:.2f}%")

# Print layout report
optimizer.print_layout_report(result)

# Visualize layout
optimizer.visualize_layout(result)
```
